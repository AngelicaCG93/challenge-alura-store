### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
import matplotlib.pyplot as plt
```

```python
tiendas = {
    'Tienda 1': tienda,
    'Tienda 2': tienda2,
    'Tienda 3': tienda3,
    'Tienda 4': tienda4
}
```

```python
ingresos = {
    'Tienda 1': tienda['Precio'].sum(),
    'Tienda 2': tienda2['Precio'].sum(),
    'Tienda 3': tienda3['Precio'].sum(),
    'Tienda 4': tienda4['Precio'].sum()
}
```

```python
ingresos
```

```python
ingresos = [df["Precio"].sum() for df in tiendas.values()]

plt.figure()
plt.bar(tiendas.keys(), ingresos)
plt.title("Ingreso total por tienda")
plt.ylabel("Ingreso total")
plt.xticks(rotation=45)
plt.show()
```

El análisis de los ingresos totales muestra que:
La tienda con mayor ingreso fue la tienda 1.
La tienda con menor ingreso fue la tienda 4.
Esto indica que la Tienda 1 tiene mayor volumen de ventas y mayor capacidad de generar rentabilidad.

# 2. Ventas por categoría

```python
print("Tienda 1")
print(tienda["Categoría del Producto"].value_counts())
print()

print("Tienda 2")
print(tienda2["Categoría del Producto"].value_counts())
print()

print("Tienda 3")
print(tienda3["Categoría del Producto"].value_counts())
print()

print("Tienda 4")
print(tienda4["Categoría del Producto"].value_counts())
```

```python
for nombre, df in tiendas.items():
    categorias = df["Categoría del Producto"].value_counts()

    plt.figure()
    plt.pie(categorias, labels=categorias.index, autopct="%1.1f%%")
    plt.title(f"Distribución de ventas por categoría - {nombre}")
    plt.show()
```

En general, se observó que la categoría más popular en la mayoría de las tiendas fue electrónicos, lo que indica una fuerte demanda en ese segmento.
Por otro lado, las categorías con menor rotación fueron artículos para el hogar, lo que podría representar una oportunidad de mejora o una menor demanda en ese mercado.

# 3. Calificación promedio de la tienda


```python
promedios = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean()
}
```

```python
promedios
```

```python
promedios = [df["Calificación"].mean() for df in tiendas.values()]

plt.figure()
plt.plot(list(tiendas.keys()), promedios, marker="o")
plt.title("Calificación promedio por tienda")
plt.ylabel("Promedio de calificación")
plt.xticks(rotation=45)
plt.show()
```

En cuanto a satisfacción del cliente:
La tienda con mejor calificación promedio fue la tienda 3.
La tienda con menor calificación fue la tienda 4.

# 4. Productos más y menos vendidos

```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in tiendas.items():
    ventas = df["Producto"].value_counts()

    print(f"\n{nombre}")
    print("Más vendido:")
    print(ventas.head(1))

    print("Menos vendido:")
    print(ventas.tail(1))
```

```python
for nombre, df in tiendas.items():
    top_productos = df["Producto"].value_counts().head(5)

    plt.figure()
    plt.barh(top_productos.index, top_productos.values)
    plt.title(f"Top 5 productos más vendidos - {nombre}")
    plt.xlabel("Cantidad vendida")
    plt.gca().invert_yaxis()
    plt.show()
```

El análisis mostró que:
Los productos más vendidos pertenecen principalmente a la categoría: microondas en la tienda 1, iniciando en programación en la tienda 2, kit de bancas en la tienda 3 y cama box en la tienda 4.
Algunos productos presentan baja rotación, lo que puede indicar poca demanda o necesidad de estrategias de marketing.
Esto permite identificar qué tipo de productos tienen mayor aceptación en cada tienda.

# 5. Envío promedio por tienda

```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

for nombre, df in tiendas.items():
    promedio_envio = df["Costo de envío"].mean()
    print(f"{nombre}: {round(promedio_envio,2)}")
```

```python
envios = [df["Costo de envío"].mean() for df in tiendas.values()]

plt.figure()
plt.bar(tiendas.keys(), envios)
plt.title("Costo de envío promedio por tienda")
plt.ylabel("Costo promedio")
plt.xticks(rotation=45)
plt.show()
```

En relación al costo de envío:
La tienda con menor costo promedio fue la tienda 4.
La tienda con mayor costo fue la tienda 1.
Dado que el cliente asume el costo de envío, un menor valor puede representar una ventaja competitiva importante.
